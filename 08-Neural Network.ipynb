```
import tensorflow as tf
n_feature = 10
n_dense_neurons = 3
```

```
import numpy as np

x = tf.placeholder(tf.float32 , (None , n_feature))
```

```
w = tf.Variable(tf.random_normal([n_feature , n_dense_neurons]))
b = tf.Variable(tf.zeros([n_dense_neurons]))
```

```
xw = tf.matmul(x,w)
```

```
z = tf.add(xw , b)
```

```
a = tf.sigmoid(z)
```

```
init = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
  sess.run(init)
  layer_out = sess.run(a, feed_dict={x:np.random.random([1, n_feature])})
```

```
print(layer_out)
```

```
t, f = 1., -1.
bias = 1.
train_in = [
            [t, t, bias],
            [t, f, bias],
            [f, t, bias],
            [f, f, bias],

]

train_out = [
             [t],
             [f],
             [f],
             [f]
]
```

```
# step(x) = {1 if x> 0 ; -1 otherwise}

def step(x):
  is_greater = tf.greater(x,0)
  as_float = tf.to_float(is_greater)
  doubled  = tf.multiply(as_float ,2)
  return tf.subtract(doubled ,1)
```

```

w=tf.Variable(tf.random_normal([3,1]))
test=tf.constant([[-1.,-1.,1]])
output=step(tf.matmul(train_in,w))
error=tf.subtract(train_out,output)
mse=tf.reduce_mean(tf.square(error))
check=tf.matmul(test,w)
```

```
delta = tf.matmul(train_in , error , transpose_a=True)
train = tf.assign(w, tf.add(w, delta))
```

```

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
err , target = 1, 0
epoch , max_epochs = 0,10
```

```
inw = sess.run(w)
print('input :' , inw)
while err > target and epoch < max_epochs:
  epoch += 1 
  err,tr = sess.run([mse , train])
  print('epoch:', epoch , 'mse: ' , err)
  print('\n')

inwo =sess.run(w)
print('out: ', inwo)
print('\n')
print('result test ' ,sess.run(check))
```
